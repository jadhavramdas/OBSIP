```python
def bi_varient_price(x):
  sns.barplot(data = df , x = x, y = 'price' , color = 'lightblue' , edgecolor = 'black')
  plt.xlabel(x.upper())
  plt.ylabel('PRICE IN MILLIONS')
  plt.title('REALTION OF PRICE AND '+str(x.upper()))
```
